```python
dataset = [
    ['milk', 'onion', 'nutmeg', 'kidney beans', 'eggs'],
    ['yogurt', 'spinach', 'cinnamon', 'black beans', 'cheese'],
    ['bread', 'tomato', 'basil', 'lentils', 'butter'],
    ['carrot', 'garlic', 'turmeric', 'chickpeas', 'cream'],
    ['potato', 'ginger', 'paprika', 'green peas', 'mayonnaise'],
    ['milk', 'cheese', 'butter', 'yogurt'],
    ['tomato', 'onion', 'basil', 'cream'],
    ['potato', 'black beans', 'lentils', 'green peas'],
    ['spinach', 'garlic', 'turmeric', 'ginger'],
    ['milk', 'yogurt', 'butter', 'cream', 'eggs'],
    ['bread', 'butter', 'cream', 'cheese'],
    ['basil', 'onion', 'tomato', 'potato'],
    ['lentils', 'chickpeas', 'black beans'],
    ['cinnamon', 'nutmeg', 'eggs'],
    ['garlic', 'onion', 'spinach'],
]
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
```

```python
te=TransactionEncoder()
te_ary=te.fit(dataset).transform(dataset)
df=pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori
frequent_itemsets = apriori(df, min_support=0.2, use_colnames=True)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
res = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
res
```
